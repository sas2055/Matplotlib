```python
import numpy as np
import matplotlib.pyplot as plt
```

## Subplots:
    - using the variable ax for single a Axes
        fig, ax = plt.subplots()

    - using the variable axs for multiple Axes
        fig, axs = plt.subplots(2, 2)

    - using tuple unpacking for multiple Axes
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

```python
# using the variable axs for multiple Axes
fig, axs = plt.subplots(2, 2)
axs
```

```python
x = np.random.random(20)
y = np.sin(x)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0,0].scatter(x,y)
axs[0,1].plot(x,y,'--o')
axs[1,0].plot(x,y,'*c')
axs[1,1].bar(['A','B'],[100,230])
```

```python
fig, axs = plt.subplots(2, 2)
axs[0,0].scatter(x,y)
axs[0,0].set_xlabel('rang1')
axs[0,0].set_ylabel('val1')
axs[0,0].set_title('Graph 1')

axs[0,1].plot(x,y,'--o')
axs[0,1].set_xlabel('rang2')
axs[0,1].set_ylabel('val2')
axs[0,1].set_title('Graph 2')

axs[1,0].plot(x,y,'*c')
axs[1,0].set_xlabel('rang3')
axs[1,0].set_ylabel('val3')
axs[1,0].set_title('Graph 3')

axs[1,1].bar(['A','B'],[100,230])
axs[1,1].set_xlabel('rang4')
axs[1,1].set_ylabel('val4')
axs[1,1].set_title('Graph 4')

plt.suptitle('Total graphs together')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].plot(x,y)
axs[1].plot(x,y,'r')
```

```python
fig, axs = plt.subplots(1, 2,sharey=True)
axs[0].plot(x,y)
axs[1].plot(x,y,'r')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1, 2,sharey=True,figsize=(6,2))
axs[0].plot(x,y)
axs[1].plot(x,y,'r')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(2, 2,figsize=(14,9))
axs[0,0].scatter(x,y)
axs[0,1].plot(x,y,'--o')
axs[1,0].plot(x,y,'*c')
axs[1,1].bar(['A','B'],[100,230])
plt.suptitle('Total graphs together')
plt.tight_layout()
```
