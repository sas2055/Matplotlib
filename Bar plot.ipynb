#### Que.1 WAP to display a bar chart of the popularity of programming Languages.

```python
import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color=['magenta'])
plt.xlabel("Languages", size=12)
plt.ylabel("Popularity", size=12)
plt.title("Popularity of Programming Language\n", size=12)
plt.xticks(x_pos, x)
plt.show()
```

#### Que.5 WAP  to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity (float value).

```python
import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]
fig, ax = plt.subplots()
r1 = ax.bar(x_pos, popularity, color='y')
plt.xlabel("Languages", size=12)
plt.ylabel("Popularity", size=12)
plt.title("Popularity of Programming Language\n", size=12)
plt.xticks(x_pos, x)
def autolabel(r2):
    for i in r2:
        height = i.get_height()
        ax.text(i.get_x() + i.get_width()/2., 1.05*height,
                '%f' % float(height),
        ha='center', va='bottom')
autolabel(r1)
plt.show()
```

#### Que.12 WAP to create bar plots with errorbars on the same figure.

```python
import numpy as np
import matplotlib.pyplot as plt
Mean = (54.74, 42.35, 67.37, 58.24, 30.25)
Std = (4, 3, 4, 1, 5)
ind = np.arange(5)     
plt.bar(ind, Mean, width=.4, yerr=Std, color='red')
plt.ylabel('Scores', size=12)
plt.xlabel('Velocity', size=12)
plt.title('Scores by Velocity\n', size=12)
plt.show()
```

#### Que.13 WAP to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying mean (integer value).

```python
import numpy as np
import matplotlib.pyplot as plt
Mean = (54.74, 42.35, 67.37, 58.24, 30.25)
Std = (4, 3, 4, 1, 5)
ind = np.arange(5)     
fig, ax = plt.subplots()
r1 = ax.bar(ind, Mean, width=0.4, color='m', 
                yerr=Std)
plt.ylabel('Scores', size=12)
plt.xlabel('Velocity', size=12)
plt.title('Scores by Velocity\n', size=12)
def autolabel(r2):
    for i in r2:
        height = i.get_height()
        ax.text(i.get_x() + i.get_width()/2., 1.05*height,
                '%d' % int(height),
        ha='center', va='bottom')
autolabel(r1)
plt.show()
```

#### Que.14 WAP to create a stacked bar plot with errorbars

```python
import numpy as np
import matplotlib.pyplot as plt
men_Means = (22, 30, 35, 35, 26)
women_Means = (25, 32, 30, 35, 29)
men_Std = (4, 3, 4, 1, 5)
women_Std = (3, 5, 2, 3, 3)
ind = np.arange(5)    
p1 = plt.bar(ind, men_Means, width=.4,
             yerr=men_Std, color='m')
p2 = plt.bar(ind, women_Means, width=.4,
        bottom=men_Means, yerr=women_Std, color='g')
plt.ylabel('Scores', size=12)
plt.xlabel('Groups', size=12)
plt.title('Scores by group\n', size=12)
plt.xticks(ind, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
plt.show()
```

#### Que.15 WAP to create stack bar plot and add label to each section.

```python
import numpy as np
import matplotlib.pyplot as plt
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4
data = [[3.40022085, 7.70632498, 6.4097905, 10.51648577,   7.5330039,
    7.1123587, 12.77792868, 3.44773477],
 [11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195,
    6.79685302, 7.24578743, 3.69371847],
 [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182,
    4.63832778,  11.16849999,   8.56883433],
 [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185,
   10.97567589, 3.98287652, 8.80552122]]
per = (np.random.randint(5, 20, (len(people), segments)))
y_pos = np.arange(len(people))
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
colors ='rgym'
patch_handles = []
left = np.zeros(len(people)) 
for i, d in enumerate(data):
    patch_handles.append(ax.barh(y_pos, d, 
      color=colors[i%len(colors)], align='center', 
      left=left))
    left += d
for j in range(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5*patch.get_width() + bl[0]
        y = 0.5*patch.get_height() + bl[1]
        ax.text(x,y, "%d%%" % (per[i,j]), ha='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Scores', size=12)
plt.show()
```

#### Que.16 WAP to add textures (black and white) to bars and wedges.

```python
import matplotlib.pyplot as plt
fig = plt.figure()
patterns = [ "|" , "\\" , "/" , "+" , "-", ".", "*", "x", "o", "O" ]

ax = fig.add_subplot(111)
for i in range(len(patterns)):
    ax.bar(i, 3, color='white', edgecolor='black', hatch=patterns[i])

plt.show()
```

#### Que.19 How To Annotate Bars in Barplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {'Name': ['Akshara', 'Sanika', 'Raj', 'Prathamesh'],
        'Marks': [55, 83, 78, 65]}
df = pd.DataFrame(data, columns=['Name', 'Marks'])
plt.figure(figsize=(8, 6))
plots = sns.barplot(x='Name', y='Marks', data=df)
plt.xlabel('Students', size=12)
plt.ylabel('Marks Secured', size=12)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {'Name': ['Akshara', 'Sanika', 'Raj', 'Prathamesh'],
        'Marks': [55, 83, 78, 65]}
df = pd.DataFrame(data, columns=['Name', 'Marks'])
plt.figure(figsize=(8, 8))
plots = sns.barplot(x='Name', y='Marks', data=df)
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')
plt.xlabel('Students', size=12)
plt.ylabel('Marks Secured', size=12)
plt.title('This is an annotated barplot\n', size=12)
plt.show()
```
