```python
plt.plot(overs,runs)
```

```python
# we need to take cummulativ sum of runs
overs = [1,2,3,4,5,6] # X
runs = [10,2,4,14,8,12] # y
```

```python
import numpy as np
run_n  = np.array(runs).cumsum()
run_n
```

```python
plt.plot(overs,run_n)
```

```python
# increase the line_width
plt.plot(overs,run_n,linewidth=10,color='yellow',label='CSK')
plt.xlabel('Overs')
plt.ylabel('Cummulative runs')
plt.title('CSK runs for 1-6 overs')
plt.legend()
```

```python
plt.plot(np.linspace(2,12,20),np.tan(np.linspace(2,12,20)))
```

### Plot a graph CSK vs RR

```python
overs
```

```python
run_n # CSK
```

```python
rr_runs = np.cumsum([0,3,15,18,20,12])
rr_runs # RR
```

```python
# increase the line_width
plt.plot(overs,run_n,linewidth=10,color='yellow',label='CSK')
plt.plot(overs,rr_runs,linewidth=10,color='magenta',label='RR')
plt.xlabel('Overs')
plt.ylabel('Cummulative runs')
plt.title('CSK VS RR 1-6 overs')
plt.legend()
```

```python
# change linestyle
plt.plot(overs,run_n,linewidth=5,color='red',label='CSK',linestyle='--')
plt.plot(overs,rr_runs,linewidth=5,color='magenta',label='RR',linestyle='--')
plt.xlabel('Overs')
plt.ylabel('Cummulative runs')
plt.title('CSK VS RR 1-6 overs')
plt.legend()
```

```python
print(dir(plt.plot))
```

```python
# marker: highligh intersection points
plt.plot(overs,run_n,marker="<",markersize=12,linestyle='--')
```

```python
# at 3 and 5 over wickets
wicket = [None, None, 16, None, 38, None]
plt.plot(overs,run_n,linestyle='--')
plt.plot(overs,wicket,linestyle=' ',marker = 'o',color='k')
```

## Bar plot-
    - It is used to plot Category vs Count/Population

```python
# Plot voting details
party = ['BJP','Shivsena','NCP','Congress','Others']
seats = [120,56,29,43,18]
```

```python
plt.bar(party,seats,color=['#ffa500','#ff4500','#0000cd','#98fb98','#000000'])
plt.xticks(rotation=90)
plt.savefig('result.jpg',dpi = 350 )
plt.show()
```

```python
plt.bar(party,seats,color='r')
plt.xticks(rotation=45)
plt.show()
```
